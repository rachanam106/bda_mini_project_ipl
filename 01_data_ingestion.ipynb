## 1.1 Structured vs Unstructured Data

| Type | Definition | Example |
|------|-----------|--------|
| **Structured** | Tabular, schema-defined, queryable | CSV, SQL tables |
| **Semi-structured** | Tags/labels but no strict schema | JSON, XML |
| **Unstructured** | No predefined format | Images, videos, tweets |

> `ipl.csv` is **structured data** — rows are matches/deliveries, columns are well-defined attributes.
>
> At ~800k+ rows it qualifies as **Big Data** under the Volume dimension of the 3Vs (Volume, Velocity, Variety).

## 1.2 Missing Value Analysis

```python
missing = df.isnull().sum()
missing_pct = (missing / len(df) * 100).round(2)

missing_report = pd.DataFrame({
    'Missing Count': missing,
    'Missing %': missing_pct
}).query('`Missing Count` > 0').sort_values('Missing %', ascending=False)

print(f'Columns with missing values: {len(missing_report)}')
missing_report
```

```python
# ── Handle missing values ─────────────────────────────────────────────────────

# Common columns that may exist — adapt as needed based on your CSV
# Strategy: fill categorical nulls with 'Unknown', numerical with 0 or median

for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].fillna('Unknown')

for col in df.select_dtypes(include='number').columns:
    df[col] = df[col].fillna(0)

print('Missing values after cleaning:')
print(df.isnull().sum().sum(), '← should be 0')
```

```python
# ── Identify key columns for analysis ────────────────────────────────────────
# Adapt column names to match YOUR ipl.csv

print('All column names:')
for i, col in enumerate(df.columns, 1):
    print(f'  {i:2}. {col}')
```
